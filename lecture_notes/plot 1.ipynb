```python
my_data =[el for el in range(1000)]
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt

my_data_2 =[el **2 for el in range(1000)]
my_data_3 = [el **3 for el in range(1000)]

plt.title("first plot", fontsize = 24)
plt.xlabel("index", fontsize = 14)
plt.ylabel("value", fontsize = 14)

plt.plot(my_data)
plt.plot(my_data_2)
plt.plot(my_data_3)
# but work only because our x values match the index values taken as X values in the graph
```

```python
%matplotlib notebook
my_data =[el **2 for el in range(10,16)]
plt.plot(my_data) 
#now the x values are wrong
```

```python
%matplotlib notebook
my_data =[el **2 for el in range(10,16)]
input_values = range(10,16)
plt.plot(input_values, my_data) #give the x and matching y values to the plot
```

```python
%matplotlib notebook

x_values = range(1,6)
y_values =[x **2 for x in x_values]

plt.scatter(x_values,y_values, s=100)
plt.savefig("my_first_plot.png") #saves the doc

plt.axis([0, 6, 0, 30]) # give the scale to use ([minX, maxX, minY, maxY])
print(list(zip(x_values, y_values))) # the zip() return tuples of the matching x and y values
```
